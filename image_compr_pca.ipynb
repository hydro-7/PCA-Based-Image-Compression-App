# **Image Compression using PCA**

```python
import numpy as np
import pandas as pd
import cv2 as cv
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img1 = cv2.cvtColor(cv2.imread('temp/im1.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img1.shape
```

```python
r, g, b = cv2.split(img1)
r = r/255
g = g/255
b = b/255
```

```python
nc = 50

pca_r = PCA(n_components=nc)
new_r = pca_r.fit_transform(r)

pca_g = PCA(n_components=nc)
new_g = pca_g.fit_transform(g)

pca_b = PCA(n_components=nc)
new_b = pca_b.fit_transform(b)
```

```python
rec_r = pca_r.inverse_transform(new_r)
rec_g = pca_g.inverse_transform(new_g)
rec_b = pca_b.inverse_transform(new_b)
```

```python
final = (cv2.merge((rec_r, rec_g, rec_b)))
```

```python
plt.imshow(final)
```
